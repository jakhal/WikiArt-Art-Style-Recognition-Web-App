```python
import torch
import torchvision
from torchvision import transforms , datasets, models
from os import listdir
import random
import torch.optim as optim
from torch.autograd import Variable
import torch.nn.functional as F 
import torch.nn as nn

from PIL import Image  
#PIL Bildverarbeitungs library from python 

normalize = transforms.Normalize(
    mean = [0.485 , 0.456 , 0.406],
    std = [0.229  , 0.224, 0.225]
)
#eine Linste vom Transformations, Funktion für die Bildverarbeitung 
#mit compose hängen wir Transformations an einander
#resize um die Bilder aller gleich groß zu machen
#normalize für die Bildqualität - nicht zu dunkel oder hell etc. 
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(256),
    transforms.ToTensor(),
    normalize])

#einen image datei importieren 
#img = Image.open("wikiartproj/aki-kuroda_untitled-1988.jpg")
#img_tensor = transform(img)  #(3,256,256)
#img_tensor.unsqueeze_(0)  #(1,3,256,256) neue Dimension mit nichts drin
#print(img_tensor)

#alle Bilder im Datei wikiartproj

#TARGET: [isAbstract, isAction_painting,isHigh_Renaissance, isMinimalism, isRococo]
train_data_list = []
train_data = []
target_list = []
#img = {x: datasets.ImageFolder(os.path.join(data_dir, x), transform[x])
                 # for x in ['train', 'val', 'test']}
files = r'C:\Users\Lilit\wikiart_ASaAI\wikiart_changed/'
files_data = listdir(files)

#zufäliges Bild wählen mit random , das heißt ändern wir die Reihenfolge der Bildklassen 

for i in range(len(files_data)): 
    f = random.choice(files_data)
    files_data.remove(f)
    img = Image.open(files + f)
    img_tensor = transform(img)  #(3,256,256) 
    train_data_list.append(img_tensor)
    train_data_list.append(img_tensor)
    isAbstract = 1 if 'Abstract' in f else 0
    isAction_painting = 1 if 'Action_painting' in f else 0
    isHigh_Renaissance = 1 if 'High_Renaissance' in f else 0
    isMinimalism = 1 if 'Minimalism' in f else 0
    isRococo = 1 if 'Rococo' in f else 0
    
    target = [isAbstract, isAction_painting,isHigh_Renaissance, isMinimalism, isRococo]
    target_list.append(target)
   # img_tensor.unsqueeze_(0)
    if len(train_data_list) >= 64:
        train_data.append((torch.stack(train_data_list), target_list))        
        train_data_list = []
        target_list = []
        print('Loaded batch', len(train_data), 'of', int(len(files_data)/64))
        print('Percentage Done:', len(train_data)/int(len(files_data)/64), '%')
       # if len(train_data) > 10:
       #     break           
#print(train_data)

class Netz ( nn.Module ):
    def _init_(self):
        super ( Netz, self ).__init__ ()
        self.conv1 = nn.Conv2d ( 1, 10, kernel_size=5 )
        self.conv2 = nn.Conv2d ( 10, 20, kernel_size=5 )
        self.conv_dropout = nn.Dropout2d ()
        self.fc1 = nn.Linear ( 320, 60 )
        self.fc2 = nn.Linear ( 60, 10 )

    def forward(self, x):
        x = self.conv1 ( x )
        x = F.max_pool2D ( x, 2 )
        x = F.relu ( x )
        x = self.conv2 ( x )
        x = self.conv_dropout ( x )
        x = F.max_pool2D ( x, 2 )
        x = F.relu ( x )
        x = x.view ( -1, 320 )
        x = F.relu ( self.fc1 ( x ) )
        x = self.fc2 ( x )
        return F.softmax ( x )

model = Netz()
model.cuda()

optimizer = optim.SGD( model.parameters (), lr=0.001, momentum=0.8 )
def train(epoch):
    model.train ()
    for batch_id, (data, target) in enumerate ( train_data ):
        data = data.cuda ()
        target = target.cuda ()
        data = Variable ( data )
        target: Variable = Variable ( target )
        optimizer.zero_grad ()
        out = model ( data )
        criterion = F.nll_loss
        loss = criterion ( out, target )
        loss.backward ()
        optimizer.step ()
        print ( 'Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format ( epoch, batch_id * len ( data ),
                                                                           len ( train_data.dataset ),
                                                                           100. * batch_id / len ( train_data ),
                                                                           loss.data[0] ) )
for epoch in range ( 1, 5 ):
    train ( epoch )
    
#test()
```
